```python
%load_ext autoreload
%autoreload 2

import sys, os
sys.path.insert(0, os.path.abspath(os.path.join(os.path.abspath(''), '../..')))

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from hydracv.midline import find_midline_midpoints, find_midline_midpoints_corrected
from hydracv.fluorescence import trace_fluo
from hydracv.utils import utils

pd.set_option('display.max_rows', 20)
```

```python
def plot(filename, path="./data/manual_labels/NGCaMP/", header='infer'):
    df = pd.read_csv(path + filename, header=header)
    cb_moments = df[df["Neural Circuit(RP, CB)"].str.contains('^(CB|CB\s)$', na=False, case=False)]['Time of Event'].values
    rp_moments = df[df["Neural Circuit(RP, CB)"].str.contains('^(RP|RP\s)', na=False, case=False)]['Time of Event'].values
    
    cb_train = utils.reformat_periods(cb_moments)
    rp_train = [utils.reformat_time(x) for x in rp_moments]
    rp_intervals = [rp_train[i] - rp_train[i-1] for i in range(1, len(rp_train))]
    plt.figure(figsize=(20,3))
    for start, end in cb_train:
        plt.hlines(0.5, start, end, linewidths=150, color='g', alpha=0.5)
    for rp in rp_train:
        plt.vlines(rp, 0.49, 0.51, color='r')
    plt.figure(figsize=(20,3))
    plt.plot(rp_intervals)
    plt.plot()
```

```python
plot("NGCaMP6_Control Manual Hydra Imaging Analysis - 0hr_Control_ngcampmov_30x4fps_50%intensity_exp2_a1.csv")
```
