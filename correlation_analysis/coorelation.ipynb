```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fluo = pd.read_csv('/home/hengji/Documents/hydra_data_analysis/trace_fluorescnece/data/NGCaMP6_50X_20fps_M1-004')
length = pd.read_csv('/home/hengji/Documents/hydra_data_analysis/find_midline/data/lengths_NGCaMP6_50X_20fps_M1_001_R1.csv')
x = np.linspace(0, len(fluo)/20, len(fluo))
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, (fluo-np.mean(fluo))/np.std(fluo),'g',label="Normalized Fluorescence")
plt.plot(x[:-1], (length-np.mean(length))/np.std(length), 'b', label="Normalized Length")
plt.legend()
plt.xlabel('Time[s]')
plt.show()
```
